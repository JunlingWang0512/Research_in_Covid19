```python

dir1 = "C:/Users/Junling_W/Desktop/Research_in_DS_course/data_arranged/covid-case/csse_covid_19_daily_reports/"
confirli = []
deathli = []
dateli = []
countryliall = []
for j in range(len(df1)):
    addre = dir1 + df1.iloc[j,0]+".csv"
    print(addre)
    df = pd.read_csv(addre,header= 0)
    df = df.fillna(0)
    countryli = []
    confirm = 0
    death = 0
    for i in range(df.shape[0]):
        if(df.iloc[i,0] is np.nan and df.iloc[i,1] is not np.nan):
            countryli.append(df.iloc[i,1])
            confirm = confirm + df.iloc[i,3]
            death = death + df.iloc[i,4]
    countryli2 = []
    for i in range(df.shape[0]):
        if(df.iloc[i,0] is not np.nan and not(df.iloc[i,1] in countryli)):
            countryli2.append(df.iloc[i,1])
            confirm = confirm + df.iloc[i,3]
            death = death + df.iloc[i,4]
    confirli.append(confirm)
    deathli.append(death)
    dateli.append(df1.iloc[j,0])
    countryliall.extend(countryli)
    countryliall.extend(countryli2)
    print(dateli[-1])
    print(confirli[-1])
```

```python
dir1 = "C:/Users/Junling_W/Desktop/Research_in_DS_course/data_arranged/covid-case/csse_covid_19_daily_reports/"
for j in range(len(df2)):
    addre = dir1 + df2.iloc[j,0]+".csv"
    print(addre)
    df = pd.read_csv(addre,header= 0)
    df.fillna(0)
    countryli = []
    confirm = 0
    death = 0
    for i in range(df.shape[0]):
        if(df.iloc[i,2] is np.nan and df.iloc[i,3] is not np.nan):
            countryli.append(df.iloc[i,3])
            confirm = confirm + df.iloc[i,7]
            death = death + df.iloc[i,8]
    countryli2 = []
    for i in range(df.shape[0]):
        if(df.iloc[i,2] is not np.nan and not(df.iloc[i,3] in countryli)):
            countryli2.append(df.iloc[i,3])
            confirm = confirm + df.iloc[i,7]
            death = death + df.iloc[i,8]
    confirli.append(confirm)
    deathli.append(death)
    dateli.append(df2.iloc[j,0])
    countryliall.extend(countryli)
    countryliall.extend(countryli2)
    print(dateli[-1])
    print(confirli[-1])
```

```python
#save worldwide cumulative data
confirmdf = pd.DataFrame(confirli, columns = ['confirm'])
deathdf = pd.DataFrame (deathli, columns = ['death'])
datedf = pd.DataFrame (dateli, columns = ['date'])
result = pd.concat([confirmdf,deathdf,datedf], axis=1)
result.to_csv('result1.csv')
#save country list
# datedf = pd.DataFrame (dateli, columns = ['date'])
#save country list
countrydf = pd.DataFrame (list(set(countryliall)), columns = ['country_or_region'])
countrydf.to_csv('countrydf.csv')
```
